```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

year = [2005, 2006, 2007, 2008, 2009, 2010]
inflation = [5, 7, 8, 12, 15, 20]
mean_inflation = 0
year_Average = 0

mean_year = np.mean(year)
mean_inflation = np.mean(inflation)
```

```python
list_of_lists = [year, inflation]

df = pd.DataFrame()

df['year'] = np.array(year)
df['inflation'] = np.array(inflation)

df
```

```python
unique_years = df['year'].unique()
for year in unique_years:
    stats.probplot(df[df['year'] == year]['inflation'], dist='norm', plot=plt)
    plt.title("Probability Plot - " + str(year))
    plt.show()
```
